```python
#Define a fn to eliminate all the brackets
def replace_func(df, col):
  char_list = ['(' , ')']

  for char in char_list:
    if char == '(':
      df[col] = df[col].apply(lambda a: str(a).replace(char,'-'))

    else:
        df[col] = df[col].apply(lambda a: str(a).replace(char,''))
```

```python
col_list = ['2019', '2018']

for col in col_list:
  replace_func(report_20, col)
```
